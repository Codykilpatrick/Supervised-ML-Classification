```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("data/classData.csv")
```

```python
#First we need to fix our category situation its a mess
#I think I want to make 1 category that has the number I.E 1001 or 0001
```

```python
df['G'] = df['G'].apply(str)
df['C'] = df['C'].apply(str)
df['B'] = df['B'].apply(str)
df['A'] = df['A'].apply(str)
```

```python
df.info()
```

```python
df['fault'] = df['G'] + df['C'] + df['B'] + df['A']
```

```python
##Now we are going to drop the first 4 columns since we dont need them anymore
```

```python
df = df.drop(['B', 'C', 'G', 'A'], axis=1)
df
```
